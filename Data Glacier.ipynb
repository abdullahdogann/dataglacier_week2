### Montly Income

* **Hypothesis:**  Is there seasonality in the number of customers using the taxi service?”

```python
df3['Date of Travel	'] = pd.to_datetime(df3['Date of Travel'],unit = 'D', origin = '1899-12-30')
```

```python
df3.drop('Date of Travel', axis = 1, inplace=True)
df3.rename(columns={'Date of Travel\t':'Date_of_Travel'},inplace=True)
df3['Date_of_Travel']=df3['Date_of_Travel'].astype(str)
```

```python
df3['Mont_and_Year'] = df3.Date_of_Travel.str.slice(0,7)
dfmontlytripcoast=df3.groupby(['Mont_and_Year','Company']).Cost_of_Trip.sum().reset_index()
dfmontlytripcoast
```

```python
plt.figure(figsize = [20,5])
sns.set_theme(style="darkgrid")
sns.lineplot(x='Mont_and_Year', y='Cost_of_Trip', hue='Company', data = dfmontlytripcoast)
plt.title('Montly Trip Income', fontsize = 30)
plt.xticks(rotation = 45);
```

* Yes, there is seasonality in the number of customers using the taxi service. As you can see, most customers have always used the taxi service during the autumn months and entering the winter.

### Age

* **Hypothesis:** Is there a difference in using the taxi service by age group?

```python
# transform age
def age_level(Age):
    if Age > 45:
        return 'old'
    elif Age > 24 and Age < 45:
        return 'midlife'
    else:
        return 'young'
```

```python
df3['age_level'] = df3.Age.apply(age_level)
```

```python
df_age = df3.groupby(['age_level','Company']).Users.count().reset_index()
plt.figure(figsize = [20,10])
sns.set_theme(style="darkgrid")
sns.barplot(x="age_level", y="Users", data=df_age)
plt.xticks(rotation = 45, fontsize=20);
plt.title('Total Users per Age Group', fontsize=30)
```

* Middle-aged people used taxi services more than younger and older people. 

## Income

* **Hypothesis:** Is there a difference in using the taxi service by income group?

```python
df3.rename(columns={'Income (USD/Month)': 'Income'}, inplace = True)
```

```python
df3['income_group'] = df3.Income.apply(lambda x: 'Poor' if x < 7500 else 'Mid Income' if x <25000 else 'Rich')
```

```python
dfincome = df3.groupby(['income_group','Company']).Users.count().reset_index()
plt.figure(figsize = [20,10])
sns.set_theme(style="darkgrid")
sns.barplot(x="income_group", y="Users", data=dfincome)
plt.xticks(rotation = 45, fontsize=20);
plt.title('Total Users per Income Group', fontsize=30)
```

* Yes; there is difference in using the taxi by age group. Middle-income people used more taxi services than poor and rich people combined.

### Population

* **Hypothesis:** Does the population in cities affect the use of taxis?

```python
df3["Population"]=df3["Population"].str.replace(",","")
df3["Users"]=df3["Users"].str.replace(",","")
df3['Population']=df3['Population'].astype(int)
df3['Users']=df3['Users'].astype(int)
```

```python
dfcity = df3[['City','Users','Population']]
dfcity
```

```python
dfusers = df3.groupby(['City']).Users.mean().reset_index()
dfpopulation = df3.groupby(['City']).Population.mean().reset_index()
dfcity = pd.merge(dfusers,dfpopulation,on='City')
dfcity
```

```python
plt.figure(figsize = [20,10])
sns.set_theme(style="darkgrid")
sns.barplot(x="City", y="Population", data=dfcity)
plt.xticks(rotation = 45)
plt.title('Population', fontsize = 30)
```

```python
plt.figure(figsize = [20,10])
sns.set_theme(style="darkgrid")
sns.barplot(x="City", y="Users", data=dfcity)
plt.xticks(rotation = 45);
plt.title('Users', fontsize = 30)
```

* **Hypothesis:**I can say that the population in cities affects the use of taxis. But there are cities like Washington to not prove it.
